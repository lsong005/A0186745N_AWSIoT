```python
from __future__ import print_function # Python 2/3 compatibility
import boto3
import json
import decimal
from boto3.dynamodb.conditions import Key, Attr
dynamodb = boto3.resource('dynamodb', 
        region_name='us-west-2', endpoint_url="http://dynamodb.us-west-2.amazonaws.com")
```

```python
table = dynamodb.Table('A0186745N')
response = table.query(
    KeyConditionExpression=Key('id').eq('FD001_1')
)
```

```python
s21=[]
for i in response['Items']:
    s21.append(i['s21'])
# print(s21)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(s21)
plt.xlabel('time')
plt.ylabel('sensordata S21')
plt.show()
```
